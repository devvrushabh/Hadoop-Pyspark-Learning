```python
from pyspark.sql import SparkSession

# Create Spark Session
spark = SparkSession.builder \
    .appName("PySparkDemo") \
    .master("local[*]") \
    .getOrCreate()
```

# 2 ways to create RDD

## 1) RDD creation from list collections

```python
data = [10,5,30,8,9,16,20,11]
rdd = spark.sparkContext.parallelize(data)
res = rdd.map(lambda x:x*x).filter(lambda x: x>100)

for i in res.collect():
  print(i)
```

## 2) RDD creation from text file

### textFile : function in spark context used to create RDD from text data

```python
path = '/content/emp.txt'

rdd = spark.sparkContext.textFile(path)

# display all rows which contains Hyderabad
# res = rdd.filter(lambda x: 'Hyderabad' in x )

# fetch only details of employ pooja
res = rdd.map(lambda x: x.split(',')).filter(lambda x: x[0] == 'pooja')

for i in res.collect():
  print(i)
```

```python
path = '/content/emp.txt'

rdd = spark.sparkContext.textFile(path)

# skip header
header = rdd.first()

res = rdd.filter(lambda x:x!=header )

# select name, sum(salary) from employ group by name

newrdd = res.map(lambda x: x.split(',')).map(lambda x: (x[0],int(x[2]))).reduceByKey(lambda x,y: x+y)

for i in newrdd.collect():
  print(i)
```

```python
data = [(1998,56000),(2000,67000),(1998,61000),(2018,26000),(1998,45000),(2000,13000)]

rdd = spark.sparkContext.parallelize(data)

## using ReduceBykey to sum the revenues by year (select year,sum(revenue) from employ group by year order by sum(revenue) )
rdd2 = rdd.reduceByKey(lambda a,b: a+b)

# Sorting the result by revenue in descending order
rdd_sorted = rdd2.sortBy(lambda x: x[1] ,ascending=False)

for i in rdd_sorted.collect():
  print(i)
```

```python
data = [(1998,56000),(2000,67000),(1998,61000),(2018,26000),(1998,45000),(2000,13000)]

rdd = spark.sparkContext.parallelize(data)

rdd_grouped= rdd.groupByKey()

# summing the values of each key
rdd_summed = rdd_grouped.mapValues(lambda x: sum(x))

for i in rdd_summed.collect():
  print(i)
```
